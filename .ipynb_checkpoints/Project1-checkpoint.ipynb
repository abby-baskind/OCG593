### Problem Description
$$
\frac{\partial f}{\partial t} + c \frac{\partial f}{\partial x} = k \frac{\partial^2 f}{\partial x^2} - \lambda f
$$

#### Scheme 1
$$ 
{f_j^{n+1}} = {f_j^{n}} (1 + \frac{c\Delta t}{\Delta x} - \frac{2\kappa \Delta t}{(\Delta x)^2} - \lambda \Delta t) + {f_{j+1}^{n}} (\frac{\kappa \Delta t}{(\Delta x)^2} - \frac{c\Delta t}{\Delta x}) + {f_{j+1}^{n}} \frac{\kappa \Delta t}{(\Delta x)^2}
$$

#### Scheme 2

$$
{f_j^{n+1}} = {f_j^{n}} (1 - \frac{c\Delta t}{\Delta x} - \frac{2\kappa \Delta t}{(\Delta x)^2} - \lambda \Delta t) + {f_{j+1}^{n}} \frac{\kappa \Delta t}{(\Delta x)^2} + {f_{j-1}^{n}} (\frac{\kappa \Delta t}{(\Delta x)^2} + \frac{c\Delta t}{\Delta x})
$$

#### Scheme 3

$$
{f_j^{n+1}} = {f_j^{n}} (1 - \frac{2\kappa \Delta t}{(\Delta x)^2} - \lambda \Delta t) + {f_{j+1}^{n}} (\frac{\kappa \Delta t}{(\Delta x)^2} - \frac{c\Delta t}{2\Delta x}) + {f_{j-1}^{n}} (\frac{\kappa \Delta t}{(\Delta x)^2} + \frac{c\Delta t}{2\Delta x})
$$

#### Boundary

$$
f(x = 0, t = t_{all}) = f(t) = \frac{t}{\tau e^{\frac{-t^2}{\tau^2}}} \\
f(x = x_{all}, t = 0) = 0 \\
f(x = L, t = t_{all}) = 0
$$


```python
# Basic packages
import numpy as np
from matplotlib import pyplot as plt
import xarray as xr
import pandas as pd
import math
import netCDF4 as nc

# DateTime packages
from matplotlib.dates import DateFormatter
from datetime import datetime, timedelta
import time
import matplotlib.dates as mdates

# Stats packages
import scipy
import bottleneck as bn
import PyCO2SYS as pyco2
import gsw
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import statsmodels.api as sm
import seaborn as sns

# Logistical packages
import requests
from importlib import reload
import warnings
warnings.filterwarnings('ignore')

import matplotlib.pyplot as plt
import cartopy.crs as ccrs
import cartopy.feature as cfeature
from matplotlib.offsetbox import AnchoredText
```

```python
dt = np.array([0.0012,0.00125,0.0013])
dx = 100
L = 1e6
N = 50
X = np.arange(0,L,dx)
u = np.zeros((N,len(X),len(dt)))
f = np.zeros((N,len(X),len(dt)))
g = np.zeros((N,len(X),len(dt)))
T = np.zeros((N, len(dt)))
tau = 15.768e6
kappa = 100
C = 0.01
h = 1.8e-9

# array for time that is Nxlen(dt)
for n,item in enumerate(T):
    for k, _ in enumerate(item):
        T[n,k] = n * dt[k]

# Boundary Conditions
for j,x in enumerate(X):
    # u(t = 0, x) = 0
    # At time 0, tracer is 0
    u[0,j,:] = 0
    f[0,j,:] = 0
    g[0,j,:] = 0
        

# Initial conditions
for k, delT in enumerate(dt):
    for n,t in enumerate(T[:,k]):
        # u(t, x = 0) = t/(Tau * e^(-t^2/Tau^2))
        # At x = 0 (j = 0), tracer is equal to t/(Tau * e^(-t^2/Tau^2))
        u[n,0,k] = t/(tau * np.exp(-t**2/tau**2))
        f[n,0,k] = t/(tau * np.exp(-t**2/tau**2))
        g[n,0,k] = t/(tau * np.exp(-t**2/tau**2))
        
        # u(t, x = L) = 0
        # At right boundary, tracer is 0
        u[n,len(X)-1,k] = 0
        f[n,len(X)-1,k] = 0
        g[n,len(X)-1,k] = 0

# Explicit differentiation scheme
for k, delT in enumerate(dt):
    for n,t in enumerate(T[:-1,k]):
        for j,x in enumerate(X[:-1]):
            # First Scheme: forward approximation for df/dx
            alpha = (C * delT)/(dx)
            gamma = (kappa * delT)/(dx**2)
            f[n+1,j,k] = f[n,j,k] * (1 + alpha - 2 * gamma - h * delT) + f[n,j-1,k] * (-alpha + gamma) + f[n,j+1,k] * gamma
            
            # Second scheme: backward approximation for df/dx
            g[n+1,j,k] = g[n,j,k] * (1 - alpha - 2 * gamma - h * delT) + g[n,j+1,k] * (gamma + alpha) + g[n,j-1,k] * gamma
            
            # Third scheme: centered approximation for df/dx
            alpha = (C * delT)/(2*dx)
            u[n+1,j,k] = u[n,j,k] * (1 - 2 * gamma - h * delT) + u[n,j+1,k] * (gamma - alpha) + u[n,j-1,k] * (gamma + alpha)
            
```

```python
fig, axarr = plt.subplots(ncols = 3, figsize = [15,5])
axarr = axarr.flatten()
nvals = np.array([0,1,5,10,49])

for k, delT in enumerate(dt):
    ax = axarr[k]
    for ni, nval in enumerate(nvals):
        # lbl = 'n = ' + str(nval) + ', t = ' + str(round(delT * nval,6))
        lbl = r't$_{' + str(nval) + '}$ = ' + str(round(delT * nval,6))
        ax.plot(X, f[nval,:,k], label = lbl, marker = 'o', ms = 2)
    ax.legend()
    ax.set_title('âˆ†t = ' + str(delT))
    ax.set_ylabel(r'$f_{x_j}^{t_n}$', fontsize = 14)
    ax.set_xlabel('x', fontsize = 14)
    
fig.suptitle('Explicit Differentation: Scheme 1')
```
